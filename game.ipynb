```python
import random
print("Welcome to Guessing Game Challenge")
print(""""The Rules are:\n1.If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"\n2.On a player's first turn, if their guess is
within 10 of the number, return "WARM!",further than 10 away from the number, return "COLD!"\n3.On all subsequent turns, if a guess is
closer to the number than the previous guess return "WARMER!",farther from the number than the previous guess, return "COLDER!"\n4.When the player's guess equals the number, tell them they've guessed correctly and how many guesses it took!""")
num =random.randint(1,100)
count=0
#print(num)
while True:
    
    guess=int(input("Enter your guess:"))
    if(guess<1 or guess>100):
        print("OUT OF BOUNDS")
        continue
        
    if(guess==num):
        print("Congratulations your guess is correct {}".format(guess))
        print("Attempts taken {}".format(count))
        break
        
    else:
        diff=abs(num-guess)
        if(diff<=10):
            print("WARM!")
            count+=1
        else:
            print("COLD!")
            count+=1
        guess1=int(input("Enter your guess:"))
       
        if(guess1<1 or guess1>100):
            print("OUT OF BOUNDS")
        
        if(guess1==num):
            print("Congratulations your guess is correct {}".format(guess))
            print("Attempts taken {}".format(count))
            break
            
        else:
            diff1=abs(guess1-num)
            if(diff1<=diff):
                print("WARMER!")
                count+=1
            else:
                print("COLDER!")
                count+=1
```
